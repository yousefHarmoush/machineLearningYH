```python
import numpy as np
import pandas as pd
import glob
path1 = r'E:\Machine-Learning\hw.xlsx'
excel_f = glob.glob(path1)
pd.set_option('display.max_rows', 91)
df1 = pd.DataFrame()
for excel_file in excel_f:
    df2 = pd.read_excel(excel_file)
    df1 = pd.concat([df1, df2], ignore_index=True)
df1.fillna(value="N/A", inplace=True)
df1.to_excel(r'E:\Machine-Learning\newhw.xlsx')
df1.head()

```
